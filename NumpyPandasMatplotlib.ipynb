```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)  #reproducibility
%matplotlib inline
```

```python
#numpy
a1 = np.array([[1, 2, 3],
              [4, 5, 6]])
print(a1.shape, a1.ndim, a1.dtype)
```

```python
#reshape & stacking
x = np.arange(12)
x_reshaped = x.reshape(3, 4)
print(x_reshaped,"\n")

v = np.array([1, 2, 3, 4])
stacked = np.vstack([x_reshaped, v.reshape(1, -1)])
print(stacked)
```

```python
print(a1[0, 0])
print(a1[0])
print(a1[:, -1])
```

```python
mask = a1 % 2 == 0
print(a1[mask])
```

```python
#mean, std, median
m_all = a1.mean()
m_rows = a1.mean(axis = 1)
s_cols = a1.std(axis = 0)
med = np.median(a1)
print(m_all, m_rows, s_cols, med)
```

```python
#pandas
students = pd.DataFrame({
    "name": ["Duong", "Inosuke", "Douma"],
    "age": [20, 20, 200],
    "major": ["AI", "CS", "CS"],
    "score": [100, 96, np.nan],
})

students.to_csv("students.csv", index = False)

df = pd.read_csv("students.csv")

display(df)
```

```python
display(df.head())
display(df.info())
display(df.describe())
```

```python
display(df[["name", "score"]])
display(df[(df["major"] == "CS") & (df["score"] >= 86)])
```

```python
display(df.groupby("major", as_index = False)["score"].mean())

display(df.isna().sum())

df_filled = df.copy()
df_filled["score"] = df_filled["score"].fillna(df_filled["score"].mean())
display(df_filled)
```

```python
#matplotlib
x = np.linspace(0, 2 * np.pi, 200)
y = np.sin(x)
plt.figure()
plt.plot(x, y)
plt.title("Inosuke Breath")
plt.xlabel("x")
plt.ylabel("sinx")
plt.show()
```

```python
n = 100
xs = np.random.randn(n)
ys = xs * 0.5 + np.random.randn(n) * 0.1
plt.figure()
plt.scatter(xs, ys)
plt.title("Random Scatter")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
data = np.random.normal(loc = 0.0, scale = 1.0, size = 1000)
plt.figure
plt.hist(data, bins = 30)
plt.title("Histogram of Normal (0, 1)")
plt.xlabel("value")
plt.ylabel("frequency")
plt.show()
```
